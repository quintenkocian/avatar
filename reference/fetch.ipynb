```python
from dotenv import load_dotenv
import os
from agents import Agent, Runner, set_default_openai_client, set_default_openai_api, set_tracing_disabled
from agents.mcp import MCPServerStdio
from openai import AsyncOpenAI
```

```python
load_dotenv(override=True)
openrouter_api_key = os.getenv("OPENROUTER_API_KEY")
```

```python
params = {
    "command": "uvx",
    "args": ["mcp-server-fetch"],
}
```

```python
client = AsyncOpenAI(
        base_url="https://openrouter.ai/api/v1",
        api_key=openrouter_api_key,
    )
set_default_openai_client(client)
set_default_openai_api("chat_completions")
set_tracing_disabled(True)
```

```python
INSTRUCTIONS = """
You are able to use your tools to browse the internet for information about Ed Donner's courses.

For information about his courses, see his personal site. The curriculum is at https://edwarddonner.com/curriculum and the specific course resources are here:
- AI Coder: https://edwarddonner.com/2026/02/17/ai-coder-vibe-coder-to-agentic-engineer/  
- AI Builder: https://edwarddonner.com/2026/01/04/ai-builder-with-n8n-create-agents-and-voice-agents/  
- AI Leadership: https://edwarddonner.com/2025/05/18/2025-ai-executive-briefing/  
- AI Engineer Core Track: https://edwarddonner.com/2024/11/13/llm-engineering-resources/  
- AI Engineer Agentic Track: https://edwarddonner.com/2025/04/21/the-complete-agentic-ai-engineering-course/  
- AI Engineer Production Track: https://edwarddonner.com/2025/09/15/ai-in-production-gen-ai-and-agentic-ai-on-aws-at-scale/  

The code repos are at llm_engineering and agents and production.

Do to the robots.txt settings at github, you might need to look up files this way:

https://raw.githubusercontent.com/ed-donner/llm_engineering/main/README.md
https://raw.githubusercontent.com/ed-donner/llm_engineering/main/week1/day1.ipynb
https://raw.githubusercontent.com/ed-donner/agents/main/1_foundations/1_lab1.ipynb

And if necessary you can look up files like this:  
https://api.github.com/repos/ed-donner/llm_engineering/git/trees/main?recursive=1

Only use your web search tool to read Ed Donner's site and the github repos for the courses, or to answer questions specifically about the courses.
Do not use it for general web browsing or unrelated questions.

"""
```

```python
async with MCPServerStdio(params, client_session_timeout_seconds=240) as server:
    agent = Agent("Fetcher", model="openai/gpt-5.4-mini", instructions=INSTRUCTIONS, mcp_servers=[server])
    response = await Runner.run(agent, "What is covered in the Week 1 Day 5 lab of llm_engineering?", max_turns=50)
    print(response.final_output)
```
